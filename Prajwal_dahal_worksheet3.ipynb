What percent of the total fare revenue came from each passenger class?

```python
total_fare=data['Fare'].sum();
print(f"total fare:{total_fare}")
fare_first_class=data[data['Pclass']==1]['Fare'].sum()
fare_second_class=data[data['Pclass']==2]['Fare'].sum()
fare_third_class=data[data['Pclass']==3]['Fare'].sum()
print(f"total fare for first class:{fare_first_class}")
print(f"total fare for second class:{fare_second_class}")
print(f"total fare for third class:{fare_third_class}")
fares_by_class=[fare_first_class,fare_second_class,fare_third_class]
proportions=[fare/total_fare*100 for fare in fares_by_class]
classes=['First Class','Second Class','Third Class']
#using zip to combine classes and proportions
for cls,prop in zip(classes,proportions):
  print(f'{cls}: {prop:.2f}%')
  #printing the class name followed by proportion formatted as percentage of two decimal places
```
