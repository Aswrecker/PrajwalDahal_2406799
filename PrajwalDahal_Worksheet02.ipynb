```python
import pandas as pd

df = pd.read_csv('/bank .csv')

df_info = df.info()

object_columns = df.select_dtypes(include=['object']).columns
unique_values = {col: df[col].unique() for col in object_columns}

null_values = df.isnull().sum()

df_numeric = df.drop(columns=object_columns)

df_numeric.to_csv('banknumericdata.csv', index=False)

df_numeric_reloaded = pd.read_csv('banknumericdata.csv')

summary_statistics = df_numeric_reloaded.describe()

df_info, unique_values, null_values, summary_statistics
#I loaded the dataset, examined its structure and missing values, identified unique categories in text columns, separated numerical data, saved it to a new file, reloaded it, and calculated summary statistics for analysis.






```

```python
df = pd.read_csv('/medical_students_dataset.csv')

df_info = df.info()

missing_values = df.isnull().sum()

if 'Age' in df.columns:
    df['Age'] = df['Age'].fillna(df['Age'].median())

if 'Gender' in df.columns:
    df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])

duplicate_values = df.duplicated().sum()

df_cleaned = df.drop_duplicates()

df_info, missing_values, duplicate_values
#I loaded the medical students' data, checked for missing values, filled missing Age with the median and Gender with the most common value, found duplicate rows, and removed them. The results show the dataset's details, missing values, and how many duplicates were found.






```
